```
import seaborn as sns
import pandas as pd
import numpy as np
```

```
tips = sns.load_dataset("tips")
print("Population size:", len(tips))
tips.head()
```

```
simple_sample = tips.sample(n=20, random_state=1)

print("Simple Random Sample")
print(simple_sample)
```

```
step = 5
systematic_sample = tips[::step]

print("Systematic Sample")
print(systematic_sample.head())
```

```
stratified_sample = tips.groupby("day").apply(lambda x: x.sample(5))

print("Stratified Sample")
print(stratified_sample)
#
```

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=tips, x="total_bill", y="tip")

plt.title("Total Bill vs Tip")
plt.show()
```

```
sns.barplot(data=tips, x="day", y="tip")

plt.title("Total Bill by Day")
plt.show()
```

```
sns.histplot(tips["total_bill"], bins=20, kde=True)

plt.title("Distribution of Total Bill")
plt.show()
#
```
